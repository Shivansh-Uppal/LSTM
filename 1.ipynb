WE NEED DATA IN MATRIX EACH ROW IS LABEL AND CORRESPONDING ROW IS FEATURE
# [[1,1:30,2,2:30,3]]  [3:30]
# [[1:30,2,2:30,3,3:30]]  [4]

```python
def df_to_X_y(df,window_size=5):
    df_as_np=df.to_numpy()
    X=[]
    y=[]
    for i in range(len(df_as_np)-window_size):       #to avoid index out of range error
        row=[[a] for a in df_as_np[i:i+5]]
        X.append(row)

        label=df_as_np[i+5]
        y.append(label)
    return np.array(X),np.array(y)
```

```python
WINDOW_SIZE=5

X,y=df_to_X_y(temp,WINDOW_SIZE)
X.shape, y.shape
```

```python
X_train,y_train=X[:100000],y[:100000]
X_valid,y_valid=X[100000:110000],y[100000:110000]
X_test,y_test=X[110000:],y[110000:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.losses import mean_squared_error
```

```python
model1=Sequential()
model1.add(InputLayer(input_shape=(WINDOW_SIZE,1)))
model1.add(LSTM(64))
model1.add(Dense(8,activation='leaky_relu'))
model1.add(Dense(1,activation='linear'))

model1.summary()
```

```python
cp=ModelCheckpoint('model1/',save_best_only=True)
model1.compile(loss=mean_squared_error,optimizer=Adam(learning_rate=0.0001),metrics=[RootMeanSquaredError()])
```

```python
model1.fit(X_train,y_train,validation_data=(X_valid,y_valid),epochs=20,callbacks=[cp])
```

```python
from tensorflow.keras.models import load_model
model1=load_model('model1/')
```

[[[1,2,3],[2,3,4]]]

```python
train_predictions=model1.predict(X_train).flatten()
train_results=pd.DataFrame(data={'y_true':y_train,'y_pred':train_predictions})
train_results
```
